```python
df_population=df_population.groupby(['geo_h3_10']).sum()
```

```python
df_target=df_target.groupby(['geo_h3_10']).sum()
df_target=df_target.drop(columns=['lat_h3','lon_h3'])
```

```python
df_companies=df_companies.fillna(0)
```

```python
df_companies=df_companies.groupby(['geo_h3_10']).sum()
```

```python
df_companies=df_companies.drop(columns=['lat','lon'])
```

```python
df_companies
```

```python
df_companies['count']=df_companies.sum(axis=1)
print(df_companies)
```

```python
df_companies=df_companies[['count']]
df_companies
```

```python
df_rating1=df_target.merge(df_population,on='geo_h3_10')
df_rating1=df_rating1.drop(columns=['lat','lon'])
df_rating1
```

```python
df_rating2=df_rating1.merge(df_companies,on='geo_h3_10')
```

```python
df_rating2
```

```python
df_rating2=df_rating2.sort_values(by=['target','atm_cnt','population'],ascending=[False,True,False])
```
